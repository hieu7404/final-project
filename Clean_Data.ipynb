# L√ÄM S·∫†CH D·ªÆ LI·ªÜU

```python
import pandas as pd
```

```python
data_clean = pd.read_csv('Data/MotChutCute.csv')
```

```python
data_clean
```

```python
data_clean.info()
```

```python
# Lo·∫°i b·ªè c·ªôt NaN

data_clean = data_clean.dropna(axis=1, how='all')
```

```python
# T√°ch reactions

def split_reactions(row):
    reaction_dict = eval(row['reactions'])
    return pd.Series(reaction_dict)
reactions = data_clean.apply(split_reactions, axis=1)
```

```python
# ƒê·ªïi reactions NaN th√†nh 0

reactions = reactions.fillna(0)
```

```python
reactions
```

```python
# N·ªëi reactions v√†o data_clean

data_clean = pd.concat([data_clean, reactions], axis=1)
data_clean
```

```python
# Chuy·ªÉn ki·ªÉu d·ªØ li·ªáu c·ªôt time sang ƒë·ªãnh d·∫°ng datetime

data_clean['time'] = pd.to_datetime(data_clean['time'])
data_clean['time']
```

```python
# T·∫°o th√™m c·ªôt ƒë·ªÉ l∆∞u th√¥ng tin v·ªÅ gi·ªù, ng√†y, th√°ng, nƒÉm

data_clean['hour'] = data_clean['time'].dt.hour
data_clean['day'] = data_clean['time'].dt.day
data_clean['month'] = data_clean['time'].dt.month
data_clean['year'] = data_clean['time'].dt.year
data_clean[['hour', 'day', 'month', 'year']]
```

```python
# L∆∞u l·∫°i d·ªØ li·ªáu ƒë√£ l√†m s·∫°ch v√† x·ª≠ l√Ω

data_clean.to_csv('Data/MotChutCute' + '_clean.csv', index=False)
```
